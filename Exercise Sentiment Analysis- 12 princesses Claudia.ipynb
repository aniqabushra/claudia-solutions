### Removing filler words (stopwords) from the list 

```python
from nltk.corpus import stopwords
eng_stopwords = stopwords.words('english')
```

```python
type(eng_stopwords)
```

```python
print(eng_stopwords)
```

```python
# Remove stop words
rm_count = 0
new_words = []  #list to hold new words

for word in txtPrincesses:
    if word not in eng_stopwords:
        new_words.append(word)
    else: rm_count += 1
```

### Shows the top 10 words in the text

```python
#the NLTK FreqDist gives a count for how often each part of the text occurs
fd_wct = FreqDist(new_words)
fd_wct
```

```python
#shows the top 10 words in the text
fd_wct.most_common(10)
```
